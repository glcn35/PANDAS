**Convert and merge json files into one Pandas.Dataframe**

- JSON JavaScript Object Notation (JavaScript Nesne Gösterimi) kısaltmasıdır.
- JSON, verileri depolamak ve farklı platformlar arasında taşıma için kullanılan bir yapıdır.
- JavaScript nesne gösterim biçiminde yazılmıştır.
- JSON programlama dilinden bağımsız olarak farklı platformlar arasında veri alışverişi için kullanılır.
- Örnek bir JSON verisi aşağıdaki gibidir:

{"kisiler": [
    { "sira": "1", "adi": "Yusuf", "soyadi": "SEZER" },
    { "sira": "2", "adi": "Ramazan", "soyadi": "SEZER" },
    { "sira": "3", "adi": "Sinan", "soyadi": "SEZER" },
    { "sira": "4", "adi": "Mehmet", "soyadi": "SEZER" }
]}

```python
import pandas as pd
import numpy as np
import glob, os, json
from pandas.io.json import json_normalize   
```

```python
data = [{'name': 'JOHN', 'age': 27}, {'name': 'HENRY2', 'age': 14}]

data2 = { 1:{'name': 'vikash', 'age': 27}, 2:{'name': 'Satyam', 'age': 14} }

data3 = [{'id': 1, 'name': "Cole Volk", 'fitness': {'height': 130, 'weight': 60}}, {'name': "Mose Reg", 'fitness': {'height': 130, 'weight': 60}}, {'id': 2, 'name': 'Faye Raker', 'fitness': {'height': 130, 'weight': 60}}]
```

```python
df1 = pd.DataFrame(data1)
df1
```

```python
df2 = pd.DataFrame(data2).T   # DataFrame'leri birleştirirken problem olmaması için satırlarla sütunları yer değiştiriyoruz.
df2
```

```python
pd.DataFrame(data3)   # Data3'ü DataFrame ile çevirdiğinizde aşağıdaki çıktıyı veriyor.
```

```python
pd.json_normalize(data3, max_level=0)    # pd.json_normalize() fonksiyonu, JSON verilerini DataFrame'e çevirmek için kullanılır.
```

```python
df3 = pd.json_normalize(data3, max_level=1)   #Data 3'ün içinde nested dictionary var. max_level parametresini 1 yaptığımızda  çıktı şağıdaki gibi oluyor.
df3
```

```python
df = pd.concat([df1,df2,df3])  # Oluşturduğumuz 3 dataframe'i birleştiriyoruz.
df
```
