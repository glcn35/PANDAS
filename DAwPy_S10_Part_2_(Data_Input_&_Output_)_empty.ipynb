# SQL Connections


```python
pip install sqlalchemy
```

```python
from sqlalchemy import create_engine  #python ile sql toolu arasında bağlantı kuracak olan argüman araç
```

```python
temp_db= create_engine("sqlite:///:memory:")  #hafızada yer oluşturuyor. ram de.şu anda bigisayarda aratsan çıkmaz. genel kullanım şekli bu
```

```python
temp_db.table_names()
```

```python
tables= pd.read_html("https://en.wikipedia.org/wiki/World_population")
```

```python
tables
```

```python
tables[6]
```

```python
tables[6].to_sql(name="populations", con=temp_db)   #bu tabloyu kendi db attık geçici olarak
```

```python
temp_db.table_names()
```

```python
df= pd.read_sql(sql="populations", con=temp_db)  #db  de kayıtlı olan tabloyu okuma
df
```

```python
pd.read_sql_query(sql="SELECT Country, Population FROM populations", con=temp_db)
```

```python
df[["Country", "Population"]]
```
