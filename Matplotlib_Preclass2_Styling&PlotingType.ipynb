```python
x = np.arange(0,30,0.2)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
plt.plot(x,y1, label="Sine")
plt.plot(x,y2, label="Cosine")
plt.legend(loc="upper right")
plt.show()
```

<h3 style="text-align: center">Bar Chart</h3>

```python
company = ["Google", "Amazon", "Microsoft", "Facebook"]
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
#xpos = [0, 1, 2, 3]
#xpos
```

```python
style.use("ggplot")
```

```python
plt.bar(company,revenue)
plt.show()
```

```python
plt.bar(xpos, revenue)
#plt.xticks(xpos, company)
plt.show()
```

```python
plt.xticks(xpos, company)
plt.ylabel("Billion Dolars")
plt.title("US Tech Stocks")
plt.bar(xpos,revenue, label="Revenue")
plt.legend()
plt.show()
```

```python
plt.xticks(xpos, company)
plt.ylabel("Billion Dolars")
plt.title("US Tech Stocks")
plt.bar(xpos,revenue, label="Revenue")
plt.bar(xpos,profit, label="Profit")
plt.legend()
plt.show()
```

```python
plt.xticks(xpos, company)
plt.ylabel("Billion Dolars")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue, label="Revenue")
plt.bar(xpos+0.2,profit, label="Profit")
plt.legend()
plt.show()
```

```python
plt.xticks(xpos, company)
plt.ylabel("Billion Dolars")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4, label="Profit")
plt.legend()
plt.show()
```

```python
plt.yticks(xpos, company)
plt.xlabel("Billion Dolars")
plt.title("US Tech Stocks")
plt.barh(xpos-0.2,revenue, height=0.4, label="Revenue")
plt.barh(xpos+0.2,profit, height=0.4, label="Profit")
plt.legend()
plt.show()
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.xlabel("Blood Sugar Range")
plt.ylabel("Total No of Patients")
plt.title("Blood Sugar Analysis")
plt.hist([blood_sugar_men, blood_sugar_women], bins=[70,100,125,150], color=["green", "orange"], label=["Men", "Women"])
plt.legend()
plt.show()
```

```python
plt.ylabel("Blood Sugar Range")
plt.xlabel("Total No of Patients")
plt.title("Blood Sugar Analysis")
plt.hist([blood_sugar_men, blood_sugar_women], bins=[70,100,125,150], color=["green", "orange"], label=["Men", "Women"], orientation="horizontal")
plt.legend()
plt.show()
```

<h3 style="text-align: center">Pie Chart</h3>

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car","Other Utilities"]
```

```python
style.use("seaborn-deep")
plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

```python
#radius
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=2)  #radius=yarcap
plt.show()
```

```python
#autopct
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%')
plt.show()
```

```python
#shadow
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%', shadow=True)  #shadow glge
plt.show()
```

```python
#explode
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%', shadow=True, explode=[0,0.2,0,0,0])  #food yularda listede 2.srada o yzeen 2.indexi yaznca o merkezdn uzaklayor
plt.show()
```

```python
#startangle
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%', shadow=True, explode=[0,0.2,0,0,0], startangle=90)     #0-360 derece aras bir say yazlrsa grafik oynatlabilir
plt.show()
```

```python
#labeldistance ve pctdistance
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%', shadow=True, explode=[0,0.2,0,0,0], labeldistance=1.1, pctdistance=0.6)
plt.show()
```
